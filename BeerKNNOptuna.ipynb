```python
def get_n_train(train, n):
    res = pd.DataFrame()
    for ind in tqdm(train["label"].unique()):
        df_label = train[train["label"] == ind]
        try:
            samples = pd.Series(df_label["Id"].unique()).sample(n=n).tolist()
        except:
            samples = pd.Series(df_label["Id"].unique()).sample(n=len(df_label["Id"].unique())).tolist()
            
        for sample in samples:
            temp_res = df_label[df_label["Id"] == sample]
            res = pd.concat([res,temp_res])

    return res
```

```python
def objective(trial):
    params = {
        'n_neighbors': trial.suggest_int('n_neighbors', 1, 100),
        'weights': trial.suggest_categorical('weights', ['uniform', 'distance']),
        'leaf_size': trial.suggest_int('leaf_size', 1, 100),
        'metric': trial.suggest_categorical('metric', ['cityblock', 'cosine', 'euclidean', 'l1', 'l2', 'manhattan', 'nan_euclidean']),
        'n_jobs': -1
    }

    clf = KNeighborsClassifier(**params)
    clf.fit(X_train, y_train)

    return clf.score(X_test, y_test)
```

```python
study = optuna.create_study(direction='maximize')
study.enqueue_trial({'n_neighbors': 8, 'weights': 'distance', 'leaf_size': 70, 'metric': 'manhattan'})
study.optimize(objective, n_trials=100)
```
